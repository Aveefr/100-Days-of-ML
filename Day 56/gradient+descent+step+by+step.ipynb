```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X, y  =make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13 )
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# OLS
from sklearn.linear_model import LinearRegression

reg = LinearRegression() 

reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), c='red')
```

```python
# assuming m = 78.35, let's apply Gradient Descent
y_pred = (X * 78.35) + 0
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), c='red', label='OLS')
plt.plot(X, y_pred, c='green', label='b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0

loss_slope = -2 * np.sum(y - m*X.ravel() -b)
loss_slope
```

```python
lr = 0.1

step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred1 = (m * X.ravel()) + b
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), c='red', label='OLS')
plt.plot(X, y_pred, c='#A3E4D7', label='b = 0')
plt.plot(X, y_pred1, c='#00a65a', label=f'b = {np.round(b,2)}')
plt.legend()
plt.show()
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), c='red', label='OLS')
plt.plot(X, y_pred, c='#A3E4D7', label='b = 0')

epochs = range(100)
m = 78.35
b = -100
lr = 0.01

for i in epochs:
    loss_slope = -2 * np.sum(y - m*X.ravel() -b)
    b = b - lr*loss_slope
    print(loss_slope, b)
    y_pred2 = (m * X.ravel()) + b
    plt.plot(X, y_pred2, c='#00a65a', alpha=((i)/100), label=f'b = {np.round(b,2)}')

plt.show()
```
