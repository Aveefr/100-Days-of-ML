```python
upper_limit = df['Height'].quantile(0.995)
lower_limit = df['Height'].quantile(0.005)
```

```python
upper_limit, lower_limit
```

```python
new_df = df[(df['Height'] > lower_limit) & (df['Height'] < upper_limit)]
```

```python
new_df.sample(5)
```

```python
new_df['Height'].describe()
```

```python
sns.kdeplot(new_df['Height'])
```

```python
sns.boxplot(new_df['Height'], orient='h')
```

```python
df['Height'] = np.where(
    df['Height'] > upper_limit, upper_limit,
    np.where(
        df['Height'] < lower_limit, lower_limit,
        df['Height']))
```

```python
df['Height'].describe()
```

```python
sns.kdeplot(df['Height'])
```

```python
sns.boxplot(df['Height'], orient='h')
```
